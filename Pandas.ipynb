```
import pandas as pd

df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/pandas/sample.csv", sep = ",", header = 0)
```

```
df
```

```
#Let's remove the dollar sign from salary
def cleanup_sal(row):
  salary = row['Salary'].replace("$", " ")
  salary = float(salary)
  return salary
```

```
#create new column for cleanup_sal 
df['cleanup_salary'] = df.apply(cleanup_sal, axis = 1)
```

```
df
```

```
#let's check the gender wise average salary 
print(df.groupby('gender')['cleanup_salary'].mean())
```

```
#Let's check job title wise average salary of female
def female_sal(row):
  if row['gender'] == "Female":
    return row['cleanup_salary']
```

```
#create new column
df['female_sal_avg'] = df.apply(female_sal, axis = 1)
```

```
#Let's check job title wise average salary of female
def male_sal(row):
  if row['gender'] == 'Male':
    return row['cleanup_salary']
```

```
df['male_sal_avg'] = df.apply(male_sal, axis = 1)
```

```
#Let's see how job title impact the salary
print(df.groupby('Job Title')['cleanup_salary', 'female_sal_avg', 'male_sal_avg'].mean())
```

```
#Let's see how job title and city impact the salary
print(df.groupby(['Job Title', 'City'])['cleanup_salary', 'female_sal_avg', 'male_sal_avg'].mean())
```
